#### Get .csv files and do data analysis here

```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# read CSV files
entropy_data = pd.read_csv('./MLP3_data/entropy_metrics.csv', skipinitialspace=True, index_col=0)
gini_data = pd.read_csv('./MLP3_data/gini_metrics.csv', skipinitialspace=True, index_col=0)
entropy_depth_data = pd.read_csv('./MLP3_data/entropy_depth_metrics.csv', skipinitialspace=True, index_col=0)
gini_depth_data = pd.read_csv('./MLP3_data/gini_depth_metrics.csv', skipinitialspace=True, index_col=0)
nb_data = pd.read_csv('./MLP3_data/nb_metrics.csv', skipinitialspace=True, index_col=0)
knn_data = pd.read_csv('./MLP3_data/knn_metrics.csv', skipinitialspace=True, index_col=0)
```

```python
print(entropy_data.head())
print(gini_data.head())
print(entropy_depth_data.head())
print(gini_depth_data.head())
print(nb_data.head())
print(knn_data.head())
```

## Decision tree

```python
plt.figure()
plt.boxplot(entropy_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
# plt.yticks([0.7, 0.8])
plt.title("Entropy tree metrics")
plt.show()
```

```python
plt.figure()
plt.boxplot(gini_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Gini tree metrics")
plt.show()
```

```python
plt.figure()
plt.boxplot(gini_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Gini tree metrics")
plt.show()
```

```python
treetype_precision = pd.DataFrame()
treetype_precision = pd.concat([entropy_data['precision'], gini_data['precision']], axis=1)

plt.figure()
plt.boxplot(treetype_precision)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model precision: Tree type")
plt.show()
```

```python
treetype_recall = pd.DataFrame()
treetype_recall = pd.concat([entropy_data['recall'], gini_data['recall']], axis=1)

plt.figure()
plt.boxplot(treetype_recall)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model recall: Tree type")
plt.show()
```

```python
treetype_accuracy = pd.DataFrame()
treetype_accuracy = pd.concat([entropy_data['accuracy'], gini_data['accuracy']], axis=1)

plt.figure()
plt.boxplot(treetype_accuracy)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model accuracy: Tree type")
plt.show()
```

```python
plt.figure()
plt.boxplot(entropy_depth_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Entropy tree metrics on depth")
plt.show()
```

```python
plt.figure()
plt.boxplot(gini_depth_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Gini tree metrics on depth")
plt.show()
```

```python
treedepth_precision = pd.DataFrame()
treedepth_precision = pd.concat([entropy_depth_data['precision'], gini_depth_data['precision']], axis=1)

plt.figure()
plt.boxplot(treedepth_precision)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model precision: Tree depth")
plt.show()
```

```python
treedepth_recall = pd.DataFrame()
treedepth_recall = pd.concat([entropy_depth_data['recall'], gini_depth_data['recall']], axis=1)

plt.figure()
plt.boxplot(treedepth_recall)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model recall: Tree depth")
plt.show()
```

```python
treedepth_accuracy = pd.DataFrame()
treedepth_accuracy = pd.concat([entropy_depth_data['accuracy'], gini_depth_data['accuracy']], axis=1)

plt.figure()
plt.boxplot(treedepth_accuracy)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model accuracy: Tree depth")
plt.show()
```

```python
treedepth_data = pd.DataFrame()
treedepth_data = pd.concat([treedepth_precision, treedepth_recall, treedepth_accuracy], axis=1)
print(treedepth_data)
treedepth_data.loc[3, :]
```

```python
x_labels = ['Entropy precision', 'Gini precision', 'Entropy recall', 'Gini recall', 'Entropy accuracy', 'Gini accuracy']
y_vals = list(treedepth_data.loc[3, :])
plt.bar(x_labels, y_vals)
plt.title("Decision tree metrics: depth of 3")
plt.rcParams["figure.figsize"] = (20,3)
plt.ylim([0.7, 0.8])
plt.show()
```

```python
x_labels = ['Entropy precision', 'Gini precision', 'Entropy recall', 'Gini recall', 'Entropy accuracy', 'Gini accuracy']
y_vals = list(treedepth_data.loc[4, :])
plt.bar(x_labels, y_vals)
plt.title("Decision tree metrics: depth of 4")
plt.rcParams["figure.figsize"] = (20,3)
plt.ylim([0.7, 0.8])
plt.show()
```

## Best models of each

```python
best_tree_data = pd.read_csv('./MLP3_data/best_tree_metrics.csv', skipinitialspace=True, index_col=0)
best_knn_data = pd.read_csv('./MLP3_data/best_knn_metrics.csv', skipinitialspace=True, index_col=0)
```

```python
best_precision = pd.DataFrame()
best_precision = pd.concat([best_tree_data['precision'], nb_data['precision'], best_knn_data['precision']], axis=1)

plt.figure()
plt.boxplot(best_precision)
plt.xticks([1, 2, 3], ['Decision Tree', 'Naive Bayes', 'kNN'])
plt.title("Model precision: Best models")
plt.show()
```

```python
best_recall = pd.DataFrame()
best_recall = pd.concat([best_tree_data['recall'], nb_data['recall'], best_knn_data['recall']], axis=1)

plt.figure()
plt.boxplot(best_recall)
plt.xticks([1, 2, 3], ['Decision Tree', 'Naive Bayes', 'kNN'])
plt.title("Model recall: Best models")
plt.show()
```

```python
best_accuracy = pd.DataFrame()
best_accuracy = pd.concat([best_tree_data['accuracy'], nb_data['accuracy'], best_knn_data['accuracy']], axis=1)

plt.figure()
plt.boxplot(best_accuracy)
plt.xticks([1, 2, 3], ['Decision Tree', 'Naive Bayes', 'kNN'])
plt.title("Model accuracy: Best models")
plt.show()
```
