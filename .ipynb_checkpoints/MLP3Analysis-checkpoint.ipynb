#### Get .csv files and do data analysis here

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read CSV files
entropy_data = pd.read_csv('./MLP3_data/entropy_metrics.csv', skipinitialspace=True, index_col=0)
gini_data = pd.read_csv('./MLP3_data/gini_metrics.csv', skipinitialspace=True, index_col=0)
entropy_depth_data = pd.read_csv('./MLP3_data/entropy_depth_metrics.csv', skipinitialspace=True, index_col=0)
gini_depth_data = pd.read_csv('./MLP3_data/gini_depth_metrics.csv', skipinitialspace=True, index_col=0)
nb_data = pd.read_csv('./MLP3_data/nb_metrics.csv', skipinitialspace=True, index_col=0)
knn_data = pd.read_csv('./MLP3_data/knn_metrics.csv', skipinitialspace=True, index_col=0)
```

```python
print(entropy_data.head())
print(gini_data.head())
print(entropy_depth_data.head())
print(gini_depth_data.head())
print(nb_data.head())
print(knn_data.head())
```

```python
plt.figure()
plt.boxplot(entropy_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
# plt.yticks([0.7, 0.8])
plt.title("Entropy tree metrics")
plt.show()
```

```python
plt.figure()
plt.boxplot(gini_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Gini tree metrics")
plt.show()
```

```python
plt.figure()
plt.boxplot(gini_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Gini tree metrics")
plt.show()
```

```python
treetype_precision = pd.DataFrame()
treetype_precision = pd.concat([entropy_data['precision'], gini_data['precision']], axis=1)

plt.figure()
plt.boxplot(treetype_precision)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model precision: Tree type")
plt.show()
```

```python
treetype_recall = pd.DataFrame()
treetype_recall = pd.concat([entropy_data['recall'], gini_data['recall']], axis=1)

plt.figure()
plt.boxplot(treetype_recall)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model recall: Tree type")
plt.show()
```

```python
treetype_accuracy = pd.DataFrame()
treetype_accuracy = pd.concat([entropy_data['accuracy'], gini_data['accuracy']], axis=1)

plt.figure()
plt.boxplot(treetype_accuracy)
plt.xticks([1, 2], ['Entropy', 'Gini'])
plt.title("Model accuracy: Tree type")
plt.show()
```

```python
plt.figure()
plt.boxplot(entropy_depth_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Entropy tree metrics on depth")
plt.show()
```

```python
plt.figure()
plt.boxplot(gini_depth_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Gini tree metrics on depth")
plt.show()
```
