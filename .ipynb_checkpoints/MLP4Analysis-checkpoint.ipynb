#### Do NN analysis and results comparison here

```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# read CSV files
model1_data = pd.read_csv('./MLP4_data/nn_model1_metrics.csv', skipinitialspace=True, index_col=0)
model2_data = pd.read_csv('./MLP4_data/nn_model2_metrics.csv', skipinitialspace=True, index_col=0)
model3_data = pd.read_csv('./MLP4_data/nn_model3_metrics.csv', skipinitialspace=True, index_col=0)
model4_data = pd.read_csv('./MLP4_data/nn_model4_metrics.csv', skipinitialspace=True, index_col=0)
model5_data = pd.read_csv('./MLP4_data/nn_model5_metrics.csv', skipinitialspace=True, index_col=0)
model6_data = pd.read_csv('./MLP4_data/nn_model6_metrics.csv', skipinitialspace=True, index_col=0)
model7_data = pd.read_csv('./MLP4_data/nn_model7_metrics.csv', skipinitialspace=True, index_col=0)
best_data = pd.read_csv('./MLP4_data/nn_bestmodel_metrics.csv', skipinitialspace=True, index_col=0)
```

```python
print(model1_data.head())
print(model2_data.head())
print(model3_data.head())
print(model4_data.head())
print(model5_data.head())
print(model6_data.head())
print(model7_data.head())
print(best_data.head())
```

## Layers

```python
plt.figure()
plt.boxplot(model1_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
plt.title("Model 1 metrics: 2 hidden layers")
plt.show()
```

```python
plt.figure()
plt.boxplot(model2_data)
plt.xticks([1, 2, 3], ['precision', 'recall', 'accuracy'])
# plt.yticks([0.7, 0.8])
plt.title("Model 2 metrics: 3 hidden layers")
plt.show()
```

```python
layer_precision = pd.DataFrame()
layer_precision = pd.concat([model1_data['precision'], model2_data['precision']], axis=1)

plt.figure()
plt.boxplot(layer_precision)
plt.xticks([1, 2], ['2 layers', '3 layers'])
plt.title("Model precision: layers")
plt.show()
```

```python
layer_recall = pd.DataFrame()
layer_recall = pd.concat([model1_data['recall'], model2_data['recall']], axis=1)

plt.figure()
plt.boxplot(layer_recall)
plt.xticks([1, 2], ['2 layers', '3 layers'])
plt.title("Model recall: layers")
plt.show()
```

```python
layer_accuracy = pd.DataFrame()
layer_accuracy = pd.concat([model1_data['accuracy'], model2_data['accuracy']], axis=1)

plt.figure()
plt.boxplot(layer_accuracy)
plt.xticks([1, 2], ['2 layers', '3 layers'])
plt.title("Model accuracy: layers")
plt.show()
```
