#### Get .csv files and do data analysis here

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read CSV files
entropy_data = pd.read_csv('./MLP3_data/entropy_metrics.csv', skipinitialspace=True, index_col=0)
gini_data = pd.read_csv('./MLP3_data/gini_metrics.csv', skipinitialspace=True, index_col=0)
entropy_depth_data = pd.read_csv('./MLP3_data/entropy_depth_metrics.csv', skipinitialspace=True, index_col=0)
gini_depth_data = pd.read_csv('./MLP3_data/gini_depth_metrics.csv', skipinitialspace=True, index_col=0)
nb_data = pd.read_csv('./MLP3_data/nb_metrics.csv', skipinitialspace=True, index_col=0)
knn_data = pd.read_csv('./MLP3_data/knn_metrics.csv', skipinitialspace=True, index_col=0)
```

```python
print(entropy_data.head())
print(gini_data.head())
print(entropy_depth_data.head())
print(gini_depth_data.head())
print(nb_data.head())
print(knn_data.head())
```
